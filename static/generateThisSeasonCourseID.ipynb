```python
import requests
import json
import os

seasonId = '202301'
res = requests.get(f'https://api.coursetable.com/api/static/catalogs/{seasonId}.json', headers={
	'Cache-Control': 'public, max-age=3600',
	'cache-control': 'public, max-age=3600',
	'Cookie': os.environ['PUBLIC_COURSETABLE_COOKIE'],
	'origin': 'https://www.coursetable.com',
	'Referer': 'https://www.coursetable.com',
	'Content-Type': 'application/json'
})

courses = res.json()
courses
```

```python
# Map across data and return course code
courseCodes = list(map(lambda course: course['same_course_id'], filter(lambda course: course['season_code'] == '202301', courses)))

# Write courseCodes to a json file called same_course_id/202301.json
with open(f'./seasons/courseIds/{seasonId}.json', 'w') as f:
	f.write(json.dumps(courseCodes))
```
